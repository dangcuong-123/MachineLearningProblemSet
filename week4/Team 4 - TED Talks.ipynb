## Chuyển đổi dấu thời gian Unix

## Xử lý ratings 

```python
import ast
```

```python
df['ratings_list'] = df.ratings.apply(lambda x: ast.literal_eval(x))

def get_name_ratings(lists):
    for i in lists:
        return i['name']

set_name_ratings = set()
for i in range(len(df)):
    set_name_ratings.add(get_name_ratings(df['ratings_list'][i]))
    
list_name_ratings = list(set_name_ratings)
```

```python
def get_count_ratings(lists, name):
    for s in lists:
        if(s['name'] == name):
            return s['count']
    return 0
for name in list_name_ratings:
    df[name] = df['ratings_list'].apply(lambda x: get_count_ratings(x, name))
```
